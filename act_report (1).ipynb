## Report: act_report


### For the twitter archive master dataset I have analysed the data and come up with various insights. 
### Some of the insights are: 
     1. What are the common dog name
     2. What are the common dog stages
     3. Is there correlation between dog name and retweet count
     4. Dog stages with the highest retweets
     5. Dog stages with the highest favorites
     6. Determining and analysing retweets and favorites counts across the              years.
     7. Is there a correlation between favorites and retweets.
     8. What is the most common  tweet source

###  The most common tweet source  is tweeter for Iphone.

### The  most common dog names are Oliver and Cooper followed by Charlie and Tucker.
### The most common dog stage is pupper followed by doggo, then puppo. Floofer is  the least common dog stage. 

### Puppo  has the most number of retweets followed by floofer then doggo. Pupper had the least number of retweets.
### Puppo  had the most number of favorite counts. Pupper had the least number of favorite counts.

### Puppo has the most number of retweets and favorite counts, while Pupper has the least number of favorites and retweet counts.

### Though pupper is the most common dog stage, it is the least retweeted and has the least favorite counts.

### There is positive correlation between favorite_counts and retweet_counts.


### There is no correlation between dog names and  favorite counts.
